```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
stats = pd.read_csv("stats.csv").dropna(axis=1)
```

```python
stats[stats['Type'] == 0][stats['nCoils'] == 4]
```

```python
plt.figure(figsize=(16, 25))
i = 1
for type in [0, 1]:
    for nCoil in [4, 8, 16]:
        print(type, nCoil)
        
        data = stats[stats['Type'] == type][stats['nCoils'] == nCoil]
        grappa_mse = data['Grappa MSE'].to_numpy()
        grappa_snr = data['Grappa SNR'].to_numpy()
        sense_mse = data['Sense MSE'].to_numpy()
        sense_snr = data['Sense SNR'].to_numpy()
        x = data['R'].to_numpy()

        plt.subplot(6, 2, i)
        plt.plot(x, grappa_mse, label='Grappa')
        plt.plot(x, sense_mse, label='Sense')
        plt.grid()
        plt.title(f"Type {type}, Coils {nCoil}, MSE")
        plt.xlabel('R');
        plt.ylabel('MSE');
        plt.legend();

        plt.subplot(6, 2, i + 1)
        plt.plot(x, grappa_snr, label='Grappa')
        plt.plot(x, sense_snr, label='Sense')
        plt.grid()
        plt.title(f"Type {type}, Coils {nCoil}, SNR")
        plt.xlabel('R');
        plt.ylabel('SNR');
        plt.legend()
        i += 2
```

```python
plt.figure(figsize=(16, 25))
i = 1
for type in [0, 1]:
    for nCoil in [4, 8, 16]:
        print(type, nCoil)
        
        data = stats[stats['Type'] == type][stats['R'] == 2]
        grappa_mse = data['Grappa MSE'].to_numpy()
        grappa_snr = data['Grappa SNR'].to_numpy()
        sense_mse = data['Sense MSE'].to_numpy()
        sense_snr = data['Sense SNR'].to_numpy()
        x = data['nCoils'].to_numpy()

        plt.subplot(6, 2, i)
        plt.plot(x, grappa_mse, label='Grappa')
        plt.plot(x, sense_mse, label='Sense')
        plt.grid()
        plt.title(f"Type {type}, R 2, MSE")
        plt.xlabel('nCoils');
        plt.ylabel('MSE');
        plt.legend();

        plt.subplot(6, 2, i + 1)
        plt.plot(x, grappa_snr, label='Grappa')
        plt.plot(x, sense_snr, label='Sense')
        plt.grid()
        plt.title(f"Type {type}, R 2, SNR")
        plt.xlabel('nCoils');
        plt.ylabel('SNR');
        plt.legend();
        i += 2
```

```python
# Surface Plot - Type 0

kind = 'SNR'
method = 'Grappa'
data = stats[stats['Type'] == 1]
x = data['R']
y = data['nCoils']
z1 = data[f'{method} {kind}']
z2 = data[f'Sense {kind}']
print(data)

fig = plt.figure(figsize=(16, 20))
ax = fig.add_subplot(projection='3d')
ax.scatter(x, y, z1, s=z1*5, label=f"{method} {kind} Type 0")
ax.plot3D(x[:5], y[:5], z1[:5], 'b--')
ax.plot3D(x[5:10], y[5:10], z1[5:10], 'b--')
ax.plot3D(x[10:15], y[10:15], z1[10:15], 'b--')
ax.scatter(x, y, z2, s=z2*5, c='r', label=f"Sense {kind} Type 0")
ax.plot3D(x[:5], y[:5], z2[:5], 'r--')
ax.plot3D(x[5:10], y[5:10], z2[5:10], 'r--')
ax.plot3D(x[10:15], y[10:15], z2[10:15], 'r--')
ax.set_xlabel('R')
ax.set_ylabel('#Coils')
ax.set_zlabel(kind)
ax.legend()
plt.show()
```
